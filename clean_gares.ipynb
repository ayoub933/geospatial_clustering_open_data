```python
import os
import requests

# Créer le dossier "data" s’il n’existe pas
os.makedirs("data", exist_ok=True)

# URL du CSV open data
url = "https://ressources.data.sncf.com/explore/dataset/liste-des-gares/download/?format=csv&timezone=Europe/Paris&use_labels_for_header=true"

# Chemin de sortie
output_path = "data/liste-des-gares.csv"

# Télécharger et enregistrer
response = requests.get(url)
with open(output_path, 'wb') as f:
    f.write(response.content)

print("✅ Fichier téléchargé automatiquement dans 'data/liste-des-gares.csv'")
```

```python
#Import des bibiliothèques
import pandas as pd
```

```python
#Création du dataframe concernant le fichier csv liste-des-gares.csv
df=pd.read_csv("liste-des-gares.csv",sep=";")
```

```python
#On garde uniquememnt les colonnes LIBELLE	VOYAGEURS	COMMUNE	DEPARTEMEN	X_WGS84	Y_WGS84
df=df.drop(columns={"CODE_UIC","FRET","CODE_LIGNE","RG_TRONCON","PK","IDRESEAU","IDGAIA"	,"X_L93"	,"Y_L93","C_GEO"	,"Geo Point",	"Geo Shape"})
```

```python
df.head()
```

```python
print(df["X_WGS84"].isnull().sum())
print(df["Y_WGS84"].isnull().sum())
```

```python
df = df.dropna(subset=["X_WGS84", "Y_WGS84"])
print(df["X_WGS84"].isnull().sum())
print(df["Y_WGS84"].isnull().sum())
```

```python
df = df.rename(columns={"VOYAGEURS": "GARE_PASSAGER"})
```

```python
df = df[df["GARE_PASSAGER"] == "O"]
```

```python
df.head()
```

```python
df_numbers=df[["X_WGS84","Y_WGS84"]]
```

```python
df_numbers.head()
```

```python
import sklearn
from sklearn.cluster import KMeans
```

```python
kmeans=KMeans(n_clusters=5)
```

```python
kmeans.fit(df_numbers)
```

```python
df_numbers.head()
```

```python
df["cluster"]=kmeans.labels_
```
